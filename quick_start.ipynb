```python
import os

%cd supersimplenet

DATASET_DIR = "./dataset"
DRIVE_DATA_PATH = "/content/drive/MyDrive/Tesi/ActiveLearning/dataset"

print("Creating local folder structure...")
os.makedirs(f"{DATASET_DIR}/train/good", exist_ok=True)
os.makedirs(f"{DATASET_DIR}/test/good", exist_ok=True)
os.makedirs(f"{DATASET_DIR}/test/reject", exist_ok=True)
os.makedirs(f"{DATASET_DIR}/ground_truth/reject", exist_ok=True)
os.makedirs(f"{DATASET_DIR}/unlabeled_pool/images", exist_ok=True)
os.makedirs(f"{DATASET_DIR}/active_pool/images", exist_ok=True)
os.makedirs(f"{DATASET_DIR}/active_pool/masks", exist_ok=True)

print("Copying training data (good samples)...")
!cp -r "{DRIVE_DATA_PATH}/train/good/." "{DATASET_DIR}/train/good/"

print("Copying test data (good and rejects)...")
!cp -r "{DRIVE_DATA_PATH}/test/good/." "{DATASET_DIR}/test/good/"
!cp -r "{DRIVE_DATA_PATH}/test/reject/." "{DATASET_DIR}/test/reject/"

if os.path.exists(f"{DRIVE_DATA_PATH}/ground_truth/reject"):
    print("Copying ground truth masks...")
    !cp -r "{DRIVE_DATA_PATH}/ground_truth/reject/." "{DATASET_DIR}/ground_truth/reject/"

print("Copying unlabeled pool for Active Learning...")
!cp -r "{DRIVE_DATA_PATH}/unlabeled_pool/images/." "{DATASET_DIR}/unlabeled_pool/images/"

print("\n[SUCCESS] Data successfully copied to local storage.")
```
